```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

NFLdata = pd.read_csv('https://raw.githubusercontent.com/bwillow1222/ds4002MI3P2/refs/heads/main/OUTPUT/csv/nfl_viewership_forecast_vs_actual.csv')
NBAdata = pd.read_csv('https://raw.githubusercontent.com/bwillow1222/ds4002MI3P2/refs/heads/main/OUTPUT/csv/nba_viewership_forecast_vs_actual.csv')
MLBdata = pd.read_csv('https://raw.githubusercontent.com/bwillow1222/ds4002MI3P2/refs/heads/main/OUTPUT/csv/mlb_viewership_forecast_vs_actual.csv')
NHLdata = pd.read_csv('https://raw.githubusercontent.com/bwillow1222/ds4002MI3P2/refs/heads/main/OUTPUT/csv/nhl_viewership_forecast_vs_actual.csv')
```

```python
# Combine the datasets, adding a 'Sport' column to differentiate them
NFLdata['Sport'] = 'NFL'
NBAdata['Sport'] = 'NBA'
MLBdata['Sport'] = 'MLB'
NHLdata['Sport'] = 'NHL'

# Combine all datasets into one
combined_data = pd.concat([NFLdata, NBAdata, MLBdata, NHLdata])

# Plot the residuals for all sports together, distinguished by color
plt.figure(figsize=(10, 6))

# Plot the residuals (percentage difference) for all sports
for sport, color in zip(['NFL', 'MLB', 'NBA', 'NHL'], ['blue', 'green', 'red', 'orange']):
    subset = combined_data[combined_data['Sport'] == sport]
    plt.scatter(subset['Year'], subset['Percentage Difference (%)'], label=sport, color=color, alpha=0.7)

# Add a horizontal line at 0
plt.axhline(0, color='black', linewidth=1)

# Add titles and labels
plt.title('Percentage Difference Residual Plot Across Sports')
plt.xlabel('Year')
plt.xticks([2020, 2021, 2022, 2023, 2024])  # Set the x-axis scale
plt.ylabel('Percentage Difference (%)')
plt.legend(title="Sport")

# Show the plot
plt.tight_layout()
plt.show()

```
